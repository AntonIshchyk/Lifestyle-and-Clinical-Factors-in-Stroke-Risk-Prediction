# Train data-balanced models

Run this notebook after `main.ipynb` has registered the prepared datasets. It trains the data-balancing comparison matrix outside the web interface and stores model files in `ai_module/data-balance-models`.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd()
if PROJECT_ROOT.name == "ai_module":
    PROJECT_ROOT = PROJECT_ROOT.parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from backend.notebook_training import (
    data_balanced_training_coverage,
    train_data_balanced_models,
    train_missing_data_balanced_models,
)
```

## Training settings

Leave `ALGORITHMS`, `DATASET_IDS`, and `BALANCING_METHODS` as `None` to run the full data-balancing comparison matrix.

```python
TRAIN_ONLY_MISSING = True
FORCE_RETRAIN = False
USE_GPU = True

ALGORITHMS = None
# Example: ALGORITHMS = ["xgboost", "lightgbm"]

DATASET_IDS = None
# Example: DATASET_IDS = ["combined_without_uncertain"]

BALANCING_METHODS = None
# Example: BALANCING_METHODS = ["random_oversampling", "weighted"]
```

```python
coverage_before = data_balanced_training_coverage()
coverage_before
```

```python
if TRAIN_ONLY_MISSING:
    result = train_missing_data_balanced_models(use_gpu=USE_GPU)
else:
    result = train_data_balanced_models(
        algorithms=ALGORITHMS,
        dataset_ids=DATASET_IDS,
        balancing_methods=BALANCING_METHODS,
        force_retrain=FORCE_RETRAIN,
        use_gpu=USE_GPU,
    )

result
```

```python
coverage_after = data_balanced_training_coverage()
coverage_after
```
